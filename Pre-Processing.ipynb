```python
d=[]
c=[]
for i in df10['sex']:
    if i == 'male':
        d.append(1)
    elif i == 'female':
        d.append(0)
for i in df10['sex']:
    if i == 'female':
        c.append(1)
    elif i == 'male':
        c.append(0)    
        
```

```python
df11=df10.copy()
```

```python
df11['male']=d
df11['female']=c
```

```python
df11
```

```python
male=0
for i in d:
    male+=i
print(male)
```

```python
female=0
for i in c:
    female+=i
print(female)
```

```python
name_df=['Male','Female']
```

```python
df_gen=[male,female]
```

```python
gender = np.array(df_gen)
mylabels = name_df

plt.title('Percentage of male and female cancer patients')
plt.pie(gender, labels = mylabels,autopct='%.2f%%')
plt.show() 
```

```python
plt.xlabel('sex') 
plt.ylabel('Gender number') 

plt.bar(name_df,df_gen)
plt.show()
```

```python
df_genarat=sum(df_gen)
df_percentages=[(i/df_genarat)*100 for i in df_gen]
```

```python

plt.bar(name_df,df_percentages)
plt.grid(color ='red',
        linestyle ='-.', linewidth = 0.8,
        alpha = 0.2)
for i, percent in enumerate(df_percentages):
    plt.text(i , percent+1, f'{percent:.2f}%', ha='left', va='center')
plt.xlabel('percentages') 
plt.ylabel('localization') 
# plt.savefig('sexlocbarplot.png')
plt.show()
```

**result**:Men are affected by 54% and women by 46%, so men are more prone to cancer
